```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

from skimage.color import rgb2gray
img=rgb2gray(cv2.imread('/content/example_EM_Image (1).tif'))
```

```python
plt.imshow((img))
```

```python
print('Mean: ',np.mean(img))
print('Standard deviation: ',np.std(img))
```

```python
plt.hist(img.ravel(), bins=256)
plt.xlabel('Pixel value')
plt.ylabel('Frequency')
plt.title('Image intensities')
plt.show()
```

The intensities shows two peaks, which likely originates from the intensities of the atomic columns and the background. There is some spread in the values.
